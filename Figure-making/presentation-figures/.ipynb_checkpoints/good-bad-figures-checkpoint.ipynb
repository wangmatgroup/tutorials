```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from IPython.display import Image
from IPython.display import set_matplotlib_formats

%matplotlib inline
```

```python
## Example: average displacement of many random walkers
## Adapted from: https://matplotlib.org/stable/gallery/lines_bars_and_markers/fill_between_alpha.html

# Fixing random state for reproducibility
np.random.seed(19680801)

Nsteps, Nwalkers = 100, 10
t = np.arange(Nsteps)

# an (Nsteps x Nwalkers) array of random walk steps
S1 = 0.004 + 0.02*np.random.randn(Nsteps, Nwalkers)
S2 = 0.002 + 0.01*np.random.randn(Nsteps, Nwalkers)

# an (Nsteps x Nwalkers) array of random walker positions
X1 = S1.cumsum(axis=0)
X2 = S2.cumsum(axis=0)

# Nsteps length arrays empirical means and standard deviations of both
# populations over time
mu1 = X1.mean(axis=1)
sigma1 = X1.std(axis=1)
mu2 = X2.mean(axis=1)
sigma2 = X2.std(axis=1)

# save the data into something readable
np.savetxt("traj1.dat", np.c_[mu1,sigma2])
np.savetxt("traj2.dat", np.c_[mu2,sigma2])
```

```python
## A better plot!

plt.rcParams.update({'font.size': 22})
plt.style.use('default')

fig, ax = plt.subplots(1,figsize=(6,4))
ax.plot(t, mu1, lw=2, label='mean population 1')
ax.plot(t, mu2, lw=2, label='mean population 2')
ax.fill_between(t, mu1+sigma1, mu1-sigma1, facecolor='C0', alpha=0.4)
ax.fill_between(t, mu2+sigma2, mu2-sigma2, facecolor='C1', alpha=0.4)

ax.legend(loc='upper left')
ax.set_xlabel('Number of steps')
ax.set_ylabel('Position (arb. units)');
plt.savefig("good-figure.png",dpi=300)
```

```python
## A better plot- with redunduncy in greyscale built in

plt.rcParams.update({'font.size': 22})
plt.style.use('default')

fig, ax = plt.subplots(1,figsize=(6,4))
ax.plot(t, mu1, lw=2, label='mean population 1')
ax.plot(t, mu2, '--', lw=2,label='mean population 2',)
ax.fill_between(t, mu1+sigma1, mu1-sigma1, facecolor='C0', alpha=0.4)
ax.fill_between(t, mu2+sigma2, mu2-sigma2, facecolor='C1', alpha=0.4)

ax.legend(loc='upper left')
ax.set_xlabel('Number of steps')
ax.set_ylabel('Position (arb. units)');
plt.savefig("good-figure-redundancy.png",dpi=300)
```

```python
## A plot with some personal preferences

plt.rcParams.update({'font.size': 18})
plt.style.use('default')

fig, ax = plt.subplots(1,figsize=(4,4))
ax.plot(t, mu1, lw=3, label='mean population 1')
ax.plot(t, mu2, lw=3, label='mean population 2')
ax.fill_between(t, mu1+sigma1, mu1-sigma1, facecolor='C0', alpha=0.4)
ax.fill_between(t, mu2+sigma2, mu2-sigma2, facecolor='C1', alpha=0.4)

ax.set_xlabel('Number of steps')
ax.set_ylabel('Position (arb. units)');
ax.grid()
plt.savefig("good-figure-2.png",dpi=300)
```

```python
## A plot for a dark background

plt.rcParams.update({'font.size': 18})
plt.style.use('dark_background')

fig, ax = plt.subplots(1,figsize=(6,4))
ax.plot(t, mu1, lw=2, color='C0', label='mean population 1')
ax.plot(t, mu2, lw=2, color='C1',label='mean population 2')
ax.fill_between(t, mu1+sigma1, mu1-sigma1, alpha=0.4)
ax.fill_between(t, mu2+sigma2, mu2-sigma2, alpha=0.4)

# increase width of spines and ticks
# change all spines
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)

# increase tick width
ax.tick_params(width=3)

ax.set_xlabel('Number of steps')
ax.set_ylabel('Position (arb. units)');
plt.tight_layout()
plt.savefig("good-figure-dark.png",dpi=300)
```
